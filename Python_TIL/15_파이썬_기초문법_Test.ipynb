- 2-1) **랜덤한 짝수 100개**로 이루어진 리스트 만들기
  - 출력 예시)   [98, 56, 16, 30, 58, 2, 72, 42, 34, 72, 84, 68, 86, 70, 82, 56, 52, 28, 90, 76, 24, 48, 8, 76, 68, 24, 96, 84, 68, 82, 34, 4, 78, 66, 36, 20, 66, 36, 10, 24, 92, 66, 8, 14, 12, 20, 36, 14, 24, 28, 32, 50, 22, 74, 70, 4, 36, 48, 14, 54, 52, 84, 4, 40, 62, 26, 6, 78, 80, 2, 66, 8, 84, 24, 22, 32, 34, 48, 56, 58, 20, 78, 48, 38, 4, 58, 24, 22, 52, 34, 80, 32, 22, 6, 100, 34, 52, 60, 36, 80]


- 2-2) 리스트의 전체 **평균 값** 출력하기
  - 출력 예시) 평균 45.88


```python
#2-1번 문제 코드작성란
import random

result = []

while True:

    if len(result) == 100:
        break
    else:
        num = random.randint(1, 100)

        if num % 2 == 0:
            result.append(num)
        else:
            continue

print(result)
```

```python
#2-2번 문제 코드작성란

result_sum = 0

for num in result:
    result_sum += num

print("평균", result_sum/len(result))  # sum(result) / len(result)
```

- 1위 : 잠

##6. 문제

- 아래 코드는 Class를 사용해서 만든 계산기 코드입니다. Class를 활용해서 아래와 같이 출력하세요
 - 5 + 2 = 7
 - 5 - 2 = 3
 - 5 * 2 = 10
 - 5 / 2 = 2.5

```python
class Calculator:
    def __init__(self, first, second):
        self.first = first
        self.second = second

    def add(self):
        result = self.first + self.second
        return result

    def sub(self):
        result = self.first - self.second
        return result

    def mul(self):
        result = self.first * self.second
        return result

    def div(self):
        result = self.first / self.second
        return result
num1 = 5
num2 = 2

calc = Calculator(num1, num2)

print(f"{num1} + {num2} = {calc.add()}")
print(f"{num1} - {num2} = {calc.sub()}")
print(f"{num1} * {num2} = {calc.mul()}")
print(f"{num1} / {num2} = {calc.div()}")
```
