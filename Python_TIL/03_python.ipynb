#### 오늘의 당첨번호

```python
import random

# random.sample(랜덤범위, 개수)
print(random.sample(range(1, 46), 6))
```

```python
import random

lotto_count = int(input("로또를 몇 개 구입하시나요? : "))
lotto_list = []

for i in range(lotto_count):
    lotto = random.sample(range(1, 46), 6)
    lotto_list.append(lotto)

for j in lotto_list:
    print(sorted(j)) # 원본 그대로 정렬 값 반환
```

```python
import random

lotto_count = int(input("로또를 몇 개 구입하시나요? : "))

for i in range(lotto_count):
    lotto = random.sample(range(1, 46), 6)
    lotto.sort() # 원본 자체를 변경하며 sort()의 값은 따로 저장되지 않음, A = sort(B) -> A : None
    print(lotto)
```
