```python
import random
import time

lunch = ["된장찌개", "피자", "제육볶음", "짜장면"]

while True:
    print(lunch)
    item = input("음식을 추가 해주세요 : ")
    if (item == "q"):
        break
    else:
        lunch.append(item)

print(lunch)

set_lunch = set(lunch)
while True:
    print(set_lunch)
    item = input("음식을 삭제해주세요 : ")
    if (item == "q"):
        break
    else:
        set_lunch = set_lunch - set([item])

print(set_lunch, "중에서 선택합니다.")

print("5")
time.sleep(1)
print("4")
time.sleep(1)
print("3")
time.sleep(1)
print("2")
time.sleep(1)
print("1")
time.sleep(1)

print(random.choice(list(set_lunch)))
```

```python
import random
import time

lunch = ["된장찌개", "피자", "제육볶음", "짜장면"]

while True:
    print(lunch)
    item = input("음식을 추가 해주세요 : ")
    if (item == "q"):
        break
    else:
        lunch.append(item)

print(lunch)

set_lunch = set(lunch) # 중복 제거와 음식 삭제를 위해
while True:
    print(set_lunch)
    item = input("음식을 삭제해주세요 : ")
    if (item == "q"):
        break
    else:
        set_lunch = set_lunch - set([item])

print(set_lunch, "중에서 선택합니다.")

for i in range(5, 0, -1):
    print("%d" %i)
    time.sleep(1)

print(random.choice(list(set_lunch)))
```

### 익명 질문 게시판

```python
total_dictionary = {}

while True:
    question = input("질문을 입력해주세요 : ")

    if (question == "q"):
        break
    else:
        total_dictionary[question] = ""

for i in total_dictionary: # dict 형식의 해당 접근은 key에 접근함
    print(i)
    answer = input("답변을 입력해주세요 :")
    total_dictionary[i] = answer

print(total_dictionary)
```

```python
total_list = []

while True:
    question = input("질문을 입력해주세요 : ")

    if (question == "q"):
        break
    else:
        total_list.append({"질문":question, "답변":""})

for i in total_list:
    print(i["질문"])
    answer = input("답변을 입력해주세요 :")
    i["답변"] = answer

print(total_list)
```
