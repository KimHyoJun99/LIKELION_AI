### 1. seoul.csv

##### 1-1. data load

```python
import csv

data = r"C:\Users\jjjun_ii\Documents\GitHub\LIKELION_AI\Data_TIL\Data_practice\data\seoul.csv"

with open(data, encoding="cp949") as infile:
    reader = csv.reader(infile)
    datas = []
    for data in reader:
        datas.append(data)
```

```python
datas
```

##### 1-2. 시각화 plot()

```python
high_temp = []
for i in range(1, len(datas)):
    if datas[i][-1] != '':
        high_temp.append(float(datas[i][-1]))
print(high_temp)
```

```python
import matplotlib.pyplot as plt

plt.plot(high_temp)
plt.show()
```

```python
date = '1907-10-1'
date_li = date.split('-')
for date in date_li:
    print(date)
```

```python
high_temp_08 = []
for i in range(1, len(datas)):
    if datas[i][-1] != '':
        if datas[i][0].split('-')[1] == "08":
            high_temp_08.append(float(datas[i][-1]))  # ''처럼 빈 리스트는 float으로 변경 X
```

```python
plt.plot(high_temp_08)
plt.show()
```

```python
high_temp_02_14 = []
for i in range(1, len(datas)):
    if datas[i][-1] != '':
        if (datas[i][0].split('-')[1] == '02') and (datas[i][0].split('-')[2] == '14'):
            high_temp_02_14.append(float(datas[i][-1]))
print(high_temp_02_14)
```

```python
plt.plot(high_temp_02_14)
plt.show()
```

```python
high_temp_over_1983 = []
low_temp_over_1983 = []
for i in range(1, len(datas)):
    if (datas[i][-1] != '') and (datas[i][-2] != ''):
        if int(datas[i][0].split('-')[0]) > 1983:
            high_temp_over_1983.append(float(datas[i][-1]))
            low_temp_over_1983.append(float(datas[i][-2]))
```

```python
plt.plot(high_temp_over_1983)
plt.plot(low_temp_over_1983)
plt.show()
```

##### 1-3. 시각화 hist()

```python
plt.hist(high_temp_over_1983, bins=100)
plt.show()
```

```python
high_temp_01 = []
high_temp_08 = []

for i in range(1, len(datas)):
    if (datas[i][-1] != ''):
        if datas[i][0].split('-')[1] == '01':
            high_temp_01.append(float(datas[i][-1]))
        if datas[i][0].split('-')[1] == '08':
            high_temp_08.append(float(datas[i][-1]))
```

```python
plt.hist(high_temp_01, bins=100, label="Jan")
plt.hist(high_temp_08, bins=100, label="Aug")
plt.legend()
plt.show()
```

```python
# 1~12월까지 저장 list 이용

month_list = [[] for _ in range(12)]  # 지정이 필요한 리스트 생성

for i in range(1, len(datas)):
    month = datas[i][0].split('-')[1]
    if (datas[i][-1] != ''):
        month_list[int(month)-1].append(float(datas[i][-1]))
print(month_list)
```

```python
# 1~12월까지 저장 dict 이용

month_dict = {}
for i in range(1, 13):
    month_dict[i] = []  # 지정해줄 값이 필요하다는 아이디어
    
for i in range(1, len(datas)):
    month = int(datas[i][0].split('-')[1])
    if datas[i][-1] != '':
        month_dict[month].append(float(datas[i][-1]))
print(month_dict)
```

```python
data = [month_dict[i] for i in range(1, 13)]
plt.boxplot(data)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

data = r"C:\Users\jjjun_ii\Documents\GitHub\LIKELION_AI\Data_TIL\Data_practice\data\seoul.csv"

with open(data, "r", encoding="cp949") as sel:
    data = csv.reader(sel)
    print(data)
    next(data)

    year = []
    month_list = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']

    for mon in month_list:
        tmp_list = []
        sel.seek(0)  # 포인터 되돌리기
        data = csv.reader(sel)  # 다시 데이터 읽기
        next(data)
        for row in data:  # 포인터를 되돌리지 않으면 csv.reader의 포인터가 끝으로 간 후 초기화되지 않아서 데이터를 읽지 못하는 문제 발생
            if row[-1] != "":
                month = row[0].split("-")[1]
                if month == mon:
                    tmp_list.append(float(row[-1]))
        year.append(tmp_list)

plt.boxplot(year)
plt.show()
```
