### 1. seoul.csv

##### 1-1. data load

```python
import csv

data = r"C:\Users\jjjun_ii\Documents\GitHub\LIKELION_AI\Data_TIL\DA\data\seoul.csv"

with open(data, encoding="cp949") as infile:
    reader = csv.reader(infile)
    datas = []
    for data in reader:
        datas.append(data)
```

```python
datas
```

##### 1-2. 시각화

```python
high_temp = []
for i in range(1, len(datas)):
    if datas[i][-1] != '':
        high_temp.append(float(datas[i][-1]))
high_temp
```

```python
import matplotlib.pyplot as plt

plt.plot(high_temp)
plt.show()
```

```python
date = '1907-10-1'
date_li = date.split('-')
for date in date_li:
    print(date)
```

```python
high_temp_08 = []
for i in range(1, len(datas)):
    if datas[i][-1] != '':
        if datas[i][0].split('-')[1] == "08":
            high_temp_08.append(float(datas[i][-1]))  # ''처럼 빈 리스트는 float으로 변경 X
```

```python
plt.plot(high_temp_08)
plt.show()
```

```python
high_temp_02_14 = []
for i in range(1, len(datas)):
    if datas[i][-1] != '':
        if (datas[i][0].split('-')[1] == '02') and (datas[i][0].split('-')[2] == '14'):
            high_temp_02_14.append(float(datas[i][-1]))
high_temp_02_14
```

```python
plt.plot(high_temp_02_14)
plt.show()
```

```python
high_temp_over_1983 = []
low_temp_over_1983 = []
for i in range(1, len(datas)):
    if datas[i][-1] != '':
        if int(datas[i][0].split('-')[0]) > 1983:
            high_temp_over_1983.append(float(datas[i][-1]))
            low_temp_over_1983.append(float(datas[i][-2]))
```

```python
plt.plot(high_temp_over_1983)
plt.show()
```

```python
plt.plot(low_temp_over_1983)
plt.show()
```
