```python
import csv

data = r"C:\Users\jjjun_ii\Documents\GitHub\LIKELION_AI\Data_TIL\DA\data\seoul.csv"

with open(data, encoding="cp949") as infile:
    reader = csv.reader(infile)
    datas = []
    for data in reader:
        datas.append(data)
```

```python
datas
```

```python
high_temp = []
for i in range(1, len(datas)):
    high_temp.append(datas[i][-1])
high_temp
```

```python
import matplotlib.pyplot as plt

plt.bar(high_temp, height=5)
plt.show()
```
