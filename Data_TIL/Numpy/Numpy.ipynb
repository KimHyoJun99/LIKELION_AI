```python
# python으로 2차원 배열 작성

import random

n = 5
m = 5

def print_list(list):
    for i in range(5):
        for k in range(5):
            print(list[i][k], end=" ")
        print()

list1 = [[0] * m for _ in range(n)]

for i in range(5):
    for k in range(5):
        list1[i][k] = random.randint(1, 300)
        
print_list(list1)
print()

for i in range(5):
    for k in range(5):
        list1[i][k] += 100

print_list(list1)
```

##### 1-2. Numpy 배열
- 넘파이는 주로 1차원, 2차원, 3차원 등의 배열을 처리하기 위한 용도로 만들어짐
- 주로 넘파이 배열을 생성할 때는 다음과 같은 함수들을 많이 사용



```python
import numpy as np
import random

pythonList = [random.randint(0, 255) for _ in range(5)]
print("* 파이썬 리스트 ==> ", pythonList)

numpyAry1 = np.array(pythonList)
print("* array(pythonList) ==> ", numpyAry1)

numpyAry2 = np.arange(5)
print("* arange(5) ==> ", numpyAry2)

numpyAry3 = np.arange(3, 8)
print("* arange(3, 8) ==> ", numpyAry3)

numpyAry4 = np.arange(0, 100, 20)
print("* arange(0, 100, 20) ==> ", numpyAry4)

numpyAry5 = np.ones(5)
print("* ones(5) ==> \n", numpyAry5)

numpyAry6 = np.ones((3, 4))
print("* ones((3, 4)) ==> \n", numpyAry6)

numpyAry7 = np.zeros(5)
print("* zeros(5) ==> ", numpyAry7)

numpyAry8 = np.zeros((3,5))
print("* zeros((3, 5)) ==> \n", numpyAry8)

numpyAry9 = np.empty(6)  # 초기화 되지 않아서 쓰레기 값이 들어있을 수 있음
print("* empty(6) ==> ", numpyAry9)

numpyAry10 = np.full(5, 33)
print("* full(5, 33) ==> ", numpyAry10)

numpyAry11 = np.identity(5)
print("* identity(5) ==> \n", numpyAry11)
```

##### 1-3. 넘파이 데이터 형식
- 넘파이 배열은 데이터 형식을 지정
- dtype으로 표현. 정수, 실수, 불리언, 복소수, 문자열 등 다양한 형식 지원
- 정수형도 8bit, 16bit, 32bit, 64bit 등을 지원. int8, int16 int32 int64 등으로 부름
- 부호가 없는 정수형도 지원하는데, 앞에 u를 붙이면 됨



```python
import numpy as np
import random

pList1 = [10, 20, 30, 40]
pList2 = [10, 20, 30, 40.0]

numpyAry1 = np.array(pList1)
print(numpyAry1, numpyAry1.dtype)

# float64, np에서 배열에는 하나의 데이터 타입만 가능, 더 높은 데이터 타입으로 통일
numpyAry2 = np.array(pList2)  
print(numpyAry2, numpyAry2.dtype)

numpyAry3 = np.array(pList1, dtype=np.float32)
print(numpyAry3, numpyAry3.dtype)

numpyAry4 = np.arange(5)
print(numpyAry4, numpyAry4.dtype)

numpyAry5 = np.ones(5)
print(numpyAry5, numpyAry5.dtype)

numpyAry6 = np.ones(5, dtype=np.uint8)
print(numpyAry6, numpyAry6.dtype)

numpyAry7 = numpyAry6.astype(np.float16)
print(numpyAry7, numpyAry7.dtype)
```

- 실행 결과를 봤을 때 배열의 부분집합을 추출했을 때 부분집합에 새로운 값이 복사되는 것이 아님
- 부분집합이 원래 배열과 메모리를 공유하는 것을 알 수 있음
- 따라서 부분집합의 값을 변경했는데 원래 배열의 값이 변경되었음




```python
# 완전히 분리된 부분집합을 만들고 싶다면 부분집합을 만들 때 복사하는 방식을 사용해야 함

ary = np.arange(10)
ary_sub2 = ary[3:7].copy()  # 복사의 방식, deep copy
print(ary_sub2)
ary_sub2[:] = 88
print(ary_sub2)
print(ary)
```

```python
# 파이썬 리스트와 넘파이 배열의 슬라이싱 비교
# 파이썬으로 슬라이싱

import random

SIZE = 5  # 원본 크기
startRow, startCol = 1, 1  # 새로운 리스트의 시작 위치
nSIZE = 3  # 새로운 리스트의 크기

myList1 = [[i+k for i in range(SIZE)] for k in range(1, SIZE**2+1, 5)]

nList = []
for i in range(startRow, startRow+nSIZE):
    rowList = []
    for k in range(startCol, startCol+nSIZE):
        rowList.append(myList1[i][k])
    nList.append(rowList)

for i in range(SIZE):
    [print("%3d" % myList1[i][k], end=' ') for k in range(SIZE)]
    print()
print()

for i in range(nSIZE):
    [print("%3d" % nList[i][k], end=' ') for k in range(nSIZE)]
    print()
```

[2차원 넘파이 배열 연결 및 축 전환]
- 2차원 넘파이 배열을 연결할 때는 np.concatenate((배열1, 배열2)) 함수 사용
- 옵션 axis를 통해 0은 세로, 1은 가로로 연결



```python
ary1 = np.zeros((2,2), dtype=np.int8)
ary2 = np.zeros((2,2), dtype=np.int8)

print(ary1)
print()
print(ary2)
print()

print(np.concatenate((ary1, ary2), axis=0))
print()
print(np.concatenate((ary1, ary2), axis=1))
```

- 넘파이 배열의 행과 열을 바꾸려면 transpose(1, 0) 또는 T 속성을 사용
- transpose(1, 0)에서 1은 1번 축을, 0은 0번 축을 가리키는데, 두 축을 바꾸기 때문에 행과 열이 바뀜
- transpose(0, 1)은 축이 바뀌지 않으므로 그대로 출력됨

```python
ary = np.array([[1, 2, 3], 
                [4, 5, 6]])
print(ary)
print()
print(ary.transpose(1, 0))
print()
print(ary.T)
print()
print(ary.transpose(0, 1))
```

### 2. Numpy 활용

##### 2-1. 수식 계산 함수


[단항 수식 함수]
- 넘파이 배열 1개를 연산
- np.abs()는 모두 양수를 반환
- np.sign()은 양수는 0, 음수는 1을 반환
- np.sqrt()는 루트값, 음수는 nan(Not a Number)로 표시



```python
import numpy as np

ary = np.array([[-1, -2], 
                [3, 4]])

print(ary)
print()
print(np.abs(ary))
print()
print(np.sign(ary))
print()
print(np.sqrt(ary))  # 경고 발생
print()
print(np.log(ary))  # 경고 발생
```

[이항 수식 함수]
- 넘파이 배열 2개를 연산
- 더하기, 빼기, 곱하기 등이 가능



```python
# np.add(배열1, 배열2)는 배열1 + 배열2와 결과가 동일함
# 함수를 사용하는 것과 직접 연산자를 사용하는 것은 차이가 없음

ary1 = np.array([[1, 2],
                 [3, 4]])
ary2 = np.array([[5, 6],
                 [7, 8]])

print(np.add(ary1, ary2))
print()
print(ary1 + ary2)
```

```python
# np.maximum(배열1, 배열2)는 두 배열 중 큰 값만 추출해서 새 배열을 반환

ary1 = np.random.randint(0, 255, size=(3,3))
ary2 = np.random.randint(0, 255, size=(3,3))

print(ary1)
print()
print(ary2)
print()
print(np.maximum(ary1, ary2))
```

```python
# np.greater(배열1, 배열2)는 '배열1 > 배열2'인 경우에 True, 그렇지 않으면 False로 구성된 새 배열을 반환

ary1 = np.random.randint(0, 255, size=(3,3))
ary2 = np.random.randint(0, 255, size=(3,3))

print(ary1)
print()
print(ary2)
print()
print(np.greater(ary1, ary2))
print()
print(ary1 > ary2)
```

```python
import numpy as np

arr = np.arange(1, 15).reshape(2, 7)

# 배열에서 10에서 20 사이의 값만을 출력
result = arr[(arr >= 10) & (arr <= 20)]
print(result)
```

##### 2-2. 기타 넘파이 기능

[조건식 표현]
- 파이썬의 'if' 문처럼 배열을 통째로 조건식을 표현할 수 있음

```python
ary1 = np.random.randint(0, 255, size=(3,3))
ary2 = np.random.randint(0, 255, size=(3,3))
condAry = np.random.choice([True, False], size=(3,3))

print(ary1)
print()
print(ary2)
print()
print(condAry)
print()
print(np.where(condAry, ary1, ary2))  # np.where(논리, True일 때 선택되는 값, False일 때 선택되는 값)
```

```python
ary = np.random.randint(-128, 127, size=(3,3))
print(ary)
print()
print(np.where(ary<0, 0, ary))
```

[통계 함수]
- 전체 배열의 합, 평균의 통계치도 구할 수 있음


```python
# 1차원 배열의 합과 평균

ary1 = np.random.randint(0, 255, size=10)
print(ary1)
print("sum ==>", ary1.sum())
print("mean ==>", ary1.mean())
```

```python
# 2차원 배열은 전체의 평균 및 합도 연산 가능
# 가로, 세로별 평균 및 덧셈 연산 가능

ary2 = np.random.randint(0, 255, size=(3,3))
print(ary2)
print()
print("sum ==>", ary2.sum())
print("axis=0, 세로의 합 ==>", ary2.sum(axis=0))  # 세로의 합
print("axis=1, 가로의 합 ==>", ary2.sum(axis=1))  # 가로의 합
```

```python
# 합이나 평균을 구할 때 조건식을 함께 사용할 수 있음

ary2 = np.random.randint(0, 255, size=(3,3))
print(ary2)
print()
print((ary2>128).sum())  # 128 이상인 항목의 개수 세기
```

```python
# 배열세서 min()으로 최솟값을, max()로 최댓값 찾기

ary2 = np.random.randint(0, 255, size=(3,3))
print(ary2)
print()
print("min ==>", ary2.min())
print("max ==>", ary2.max())
```

[정렬]
- np.sort() 함수는 배열을 정렬
- 기본적으로 오름차순, 내림차순으로 정렬하려면 정렬 결과에 [::-1]을 붙이면 됨

```python
ary = np.random.randint(0, 255, size=10)
print(ary)
print()
print(np.sort(ary))
print()
print(np.sort(ary)[::-1])
```

```python
arr = np.arange(12).reshape(3, 4)
print(arr)
print()
print(arr[0:3:-1])
print()
print(arr[2:0:-1])  # 2차원 배열에서 행을 마지막부터 출력
print()
print(arr[1:,2::-1])  # 2차원 배열에서 1행부터 마지막까지, 2열부터 마지막까지 열을 반대로 출력
print()
print(arr[1:,3::-1])
print()
print(arr[1:][::-1])  # 2차원 배열이기 때문에 행이 뒤바뀌는 결과
print()
print(arr[1::-1, 2::-1])  # 2차원 배열에서 행과 열을 모두 반대로 출력
```

[중복값 제거와 공통값 추출]

- 두 배열 중에서 공통된 항목만 추출하려면 np.intersect1d(배열1, 배열2)
- 공통된 항목들이 정렬되어 반환됨

```python
ary1 = np.random.randint(0, 9, size=10)
print(ary1)
print()
ary2 = np.random.randint(0, 9, size=10)
print(ary2)
print()
print(np.intersect1d(ary1, ary2))
```

```python
# 작업을 진행하다가, Ctrl + z를 누르면 직전 값으로 되돌리는 과정을 단순화해서 구현

import numpy as np

SIZE = 5  # 원본 크기

# 넘파이 2차원 배열 생성
imageAry = np.random.randint(0, 255, size=(SIZE, SIZE))
print("### 1. 원본 ###")
print(imageAry)
np.save("source", imageAry)

# 10 증가 후 저장
imageAry += 10
print("### 2. 10 증가 ###")
print(imageAry)
np.save("result", imageAry)

# 흑백 처리 후 저장
imageAry = np.where(imageAry<128, 0, 255)
print("### 3. 흑백 처리 ###")
print(imageAry)
np.save("result2", imageAry)

# 반전 처리 후 저장
imageAry = 255 - imageAry
print("### 4. 반전 처리 ###")
print(imageAry)
np.save("result3", imageAry)

# 복구 1
imageAry = np.load("result2.npy")
print("### 복구1: result2.npy ###")
print(imageAry)

# 복구 2
imageAry = np.load("result.npy")
print("### 복구2: result.npy ###")
print(imageAry)

# 복구 3
imageAry = np.load("source.npy")
print("### 복구3(원본): source.npy ###")
print(imageAry)
```
