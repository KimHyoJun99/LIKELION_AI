```python
# 여러 개의 선 그리기

x_data = ["1st", "2nd", "3rd", "4th", "5th"]
y1_data = [90, 28, 75, 58, 78]
y2_data = [80, 80, 50, 40, 90]
y3_data = [40, 50, 90, 90, 60]
plt.plot(x_data, y1_data, "r-o", x_data, y2_data, "g:x", x_data, y3_data, "b--p")
plt.show()
```

```python
# 여러 막대그래프 그리기

import numpy as np
import matplotlib.pyplot as plt

x_data = ["1st", "2nd", "3rd", "4th", "5th"]
x_value = np.array([1, 2, 3, 4, 5])
y1_data = np.array([90, 28, 75, 58, 78])
y2_data = np.array([80, 80, 50, 40, 90])
y3_data = np.array([40, 50, 90, 90, 60])

plt.bar(x_value, y1_data, color="red", width=0.1, label="DaHyun")
plt.bar(x_value+0.2, y2_data, color="green", width=0.1, label="JungYoun")
plt.bar(x_value+0.4, y3_data, color="blue", width=0.1, label="Momo")
plt.xticks(x_value+0.2, x_data)  # x축 눈금을 나타내고 설정
plt.legend(loc="upper right")  # 범례
plt.show()
```

```python
plt.barh(x_value, y1_data, color="red", height=0.1, label="Dahyun")
plt.barh(x_value+0.2, y2_data, color="green", height=0.1, label="JungYoun")
plt.barh(x_value+0.4, y3_data, color="blue", height=0.1, label="Momo")
plt.yticks(x_value+0.2, x_data)
plt.legend(loc="upper right")
plt.show()
```

```python
# 산점도

import numpy as np
import matplotlib.pyplot as plt

SIZE = 30
x_value = np.random.rand(SIZE)  # 0이상 1미만의 범위에서 30개 생성
y_value = np.random.rand(SIZE)
sizeAry = (50 * np.random.rand(SIZE)) ** 2
colorAry = np.random.rand(SIZE)

plt.scatter(x_value, y_value, s=sizeAry, c=colorAry, alpha=0.5, cmap="spring")
plt.colorbar()
plt.show()
```

### 2. Pandas와 Matplotlib 활용

##### 2-1. CSV와 엑셀 데이터 처리

```python
# CSV 파일을 데이터프레임으로 읽어서 데이터를 정렬한 후 출력

import pandas as pd

filename = "C:\\Users\\jjjun_ii\\Documents\\GitHub\\LIKELION_AI\\Data_TIL\\Matplotlib\\data\\singer2.csv"
df = pd.read_csv(filename, index_col=None, encoding="cp949")

df_sort1 = df.sort_values(by=["평균 키"], axis=0)
print(df_sort1)
```

```python
# 엑셀 파일의 워크시트 여러 개를 읽어서 하나의 데이터프레임으로

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

inFilename = "C:\\Users\\jjjun_ii\\Documents\\GitHub\\LIKELION_AI\\Data_TIL\\Matplotlib\\data\\singer.xlsx"
outFilename = "C:\\Users\\jjjun_ii\\Documents\\GitHub\\LIKELION_AI\\Data_TIL\\Matplotlib\\data\\singer_over6.xlsx"

df = pd.read_excel(inFilename)
print(df)  # 그냥 읽어오면 첫번째 시트만 저장됨
print()
df_senior = pd.read_excel(inFilename, "senior", index_col=None)  # 시트 이름 명시
df_junior = pd.read_excel(inFilename, "junior", index_col=None)

df_singer = pd.concat([df_senior, df_junior])
df_singer_over6 = df_singer[df_singer["인원"]>=6]
df_singer_over6 = df_singer_over6.sort_values(by=["인원"], axis=0, ascending=False)
print(df_singer_over6)
print()
df_singer_over6 = df_singer_over6.loc[:, ["아이디", "이름", "인원", "유튜브 조회수"]]  # 모든 행, 리스트 내의 칼럼
print(df_singer_over6)

x_data = df_singer_over6["아이디"]
y_data = df_singer_over6["인원"]
colorAry = [np.random.choice(["red", "green", "blue", "brown", "gold", "lime", 
                              "aqua", "magenta", "purple"]) for _ in range(len(x_data))]
plt.bar(x_data, y_data, color=colorAry)
plt.show()

writer = pd.ExcelWriter(outFilename)
df_singer_over6.to_excel(writer, sheet_name="singer", index=False)
writer.close()
print("Save. OK~")
```

```python
# 유튜브 조회수로 정렬 후 산점도로 표현
import pandas as pd
import matplotlib.pyplot as plt

inFilename = "C:\\Users\\jjjun_ii\\Documents\\GitHub\\LIKELION_AI\\Data_TIL\\Matplotlib\\data\\singer.xlsx"
dataframes = pd.read_excel(inFilename, sheet_name=None)

for df in dataframes:
    print(df)
df1 = dataframes["senior"]
df2 = dataframes["junior"]
df3 = pd.concat([df1, df2], ignore_index=True)
df3 = df3.sort_values(by=["유튜브 조회수"], ascending=False)
df3

x_data = df3["아이디"]
y_data = df3["유튜브 조회수"]
sizeAry = np.sqrt(df3["유튜브 조회수"])

plt.scatter(x_data, y_data, s=sizeAry, c=df3["유튜브 조회수"], alpha=0.5, cmap="spring")
plt.colorbar()
plt.show()
```
