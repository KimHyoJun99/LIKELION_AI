### 2. seaborn

```python
#   
data = r"C:\Users\jjjun_ii\Documents\GitHub\LIKELION_AI\Data_TIL\Visualization\data\health_screenings_2020_1000ea.xlsx"
df_health = pd.read_excel(data)

df_health_selected = df_health[["gender", "height", "weight", "waist", "drinking", "smoking"]]
df_health_selected.loc[df_health_selected['gender']==1,['gender']]='Male'
df_health_selected.loc[df_health_selected['gender']==2,['gender']]='Female'
df_health_selected.loc[df_health_selected['drinking']==0,['drinking']]='Non-drinking'
df_health_selected.loc[df_health_selected['drinking']==1,['drinking']]='Drinking'
df_health_selected.loc[(df_health_selected['smoking']==1) | (df_health_selected['smoking']==2),['smoking']]='Non-smoking'
df_health_selected.loc[df_health_selected['smoking']==3,['smoking']]='Smoking'
df_health_selected.head()
```

```python
drinking = df_health_selected.groupby(["gender", "drinking"])["drinking"].count()
smoking = df_health_selected.groupby(["gender", "smoking"])["smoking"].count()
drinking = drinking.to_frame(name="count")  # Series Dataframe
smoking = smoking.to_frame(name="count")
drinking = drinking.reset_index()
smoking = smoking.reset_index()
```

```python
drinking
```

```python
import seaborn as sns

fig = plt.figure(figsize=(17, 6))

area1 = fig.add_subplot(1, 2, 1)
area2 = fig.add_subplot(1, 2, 2)

ax1 = sns.barplot(data=drinking, x="gender", y="count", hue="drinking", ax=area1)
ax2 = sns.barplot(data=smoking, x="gender", y="count", hue="smoking", ax=area2)

fig.suptitle("Seaborn Bar Graph", fontweight="bold")
area1.set_title("Drinking Type")
area2.set_title("Smoking Type")

plt.show()
```

```python
fig = plt.figure(figsize=(17, 6))
fig.suptitle("Seaborn Bar Graph", fontweight="bold")
index = np.arange(4)

fig.add_subplot(1, 2, 1)  #  ,  ,   

plt.bar(index, drinking["count"])
plt.title("Drinking Type")
plt.ylabel("Count")
plt.xticks(index, ["Non-drinking(Female)", "Drinking(Female)", "Non-dringking(Male)", "Drinking(Male)"])
fig.add_subplot(1, 2, 2)

plt.bar(index, smoking["count"])
plt.title("Smoking Type")
plt.ylabel("Count")
plt.xticks(index, ["Non-smoking(Female)", "Smoking(Female)", "Non-smoking(Male)", "Smoking(Male)"])

plt.show()
```

##### 4-2. Data Preprocessing

```python
df_city["dt"] = pd.to_datetime(df_city["dt"])
df_city.info()
```

```python
df_city["AverageTemperature"] = df_city['AverageTemperature'].fillna(df_city['AverageTemperature'].mean())
df_city.info()
```

```python
df_city["year"] = df_city["dt"].dt.year
df_city["month"] = df_city["dt"].dt.month
df_city["day"] = df_city["dt"].dt.day
df_city.head()
```

```python
df_city["Latitude_num"] = df_city["Latitude"].apply(lambda x: float(x[:-1]) if x[-1] == "N" else -float(x[:-1]))
df_city["Longitude_num"] = df_city["Longitude"].apply(lambda x: float(x[:-1]) if x[-1] == "E" else -float(x[:-1]))
df_city.head()
```

##### 4-3. Data Visualization

```python
df_city_abidjan = df_city[df_city["City"] == "Abidjan"]
```

```python
plt.plot(df_city_abidjan["dt"], df_city_abidjan["AverageTemperature"])
```

```python
df_latitude_group = df_city.groupby("Latitude_num")["AverageTemperature"].mean()
df_latitude_group = df_latitude_group.to_frame(name="latitude_mean_temperature")
df_latitude_group.reset_index(inplace=True)
df_latitude_group.head()
```

```python
df_latitude_group.plot(x="Latitude_num", y="latitude_mean_temperature")
```
