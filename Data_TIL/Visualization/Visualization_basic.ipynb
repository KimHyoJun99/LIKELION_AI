```python
import csv
import matplotlib.pyplot as plt

data = r"C:\Users\jjjun_ii\Documents\GitHub\LIKELION_AI\Data_TIL\Data_practice\data\seoul.csv"

with open(data, "r", encoding="cp949") as sel:
    data = csv.reader(sel)
    print(data)
    next(data)

    year = []
    month_list = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']

    for mon in month_list:
        tmp_list = []
        sel.seek(0)  # 포인터 되돌리기
        data = csv.reader(sel)  # 다시 데이터 읽기
        next(data)
        for row in data:  # 포인터를 되돌리지 않으면 csv.reader의 포인터가 끝으로 간 후 초기화되지 않아서 데이터를 읽지 못하는 문제 발생
            if row[-1] != "":
                month = row[0].split("-")[1]
                if month == mon:
                    tmp_list.append(float(row[-1]))
        year.append(tmp_list)

plt.boxplot(year)
plt.show()
```

```python
import pandas as pd

data = r"C:\Users\jjjun_ii\Documents\GitHub\LIKELION_AI\Data_TIL\Data_practice\data\dust.xlsx"
df = pd.read_excel(data)
df.head()
```

```python
df.columns = ["date", "so2", "co", "o3", "no2", "PM10", "PM2.5"]
df.head()
```

```python
df["date"] = df["date"].str[:10]
df.head()
```

```python
df["date"] = pd.to_datetime(df["date"])
df.dtypes
```

```python
df["year"] = df["date"].dt.year
df["month"] = df["date"].dt.month
df["day"] = df["date"].dt.day

df.head()
```

```python
# 정렬된 dataframe 새로 생성

df_sorted = pd.DataFrame({"date": df["date"], "year": df["year"], 
                          "month": df["month"], "day": df["day"], 
                          "so2": df["so2"], "co": df["co"], 
                          "o3": df["o3"], "no2": df["no2"], 
                          "PM10": df["PM10"], "PM2.5": df["PM2.5"]})
df_sorted.head()
```

```python
# 칼럼 순서 바꾸는 간단한 방법

df = df[["date", "year", "month", "day", "so2", "co", "o3", "no2", "PM10", "PM2.5"]]
df.head()
```

```python
df.isnull().sum()
```

```python
df[df["PM10"].isna()]  # 279, 280, 689, 690은 날짜를 제외하고 모두 NaN
```

```python
# 시계열 분석이므로 null 값을 이전 시간의 값을 기준으로 채워줌
# 결측값을 앞 방향 혹은 뒷 방향으로 채우기

df = df.fillna(method="pad")  # method = pad: 이전 행의 값으로 대체
df.isnull().sum()  # 채워지지 않은 데이터는 이전 값이 없기 떄문에 => 0행
```
