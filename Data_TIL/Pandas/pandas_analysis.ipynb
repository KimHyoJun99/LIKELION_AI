### 1. Iris 데이터

##### 1-1. data load

```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np

iris = load_iris()
iris
```

```python
import seaborn as sns

df_sns = sns.load_dataset("iris")
df_sns
```

```python
df = pd.DataFrame(data = np.c_[iris["data"], iris["target"]],
                  columns=iris["feature_names"]+["target"])  # np.c_: 열로 합치기, columns는 파이썬 리스트 연산
df
```

```python
iris["target"] = iris["target"].map({0: "setosa", 1: "versicolor", 2: "virginica"})  # 0, 1, 2 mapping
iris  # numpy 배열로 되어 있어서 map 사용 X
```

```python
class_mapping = {0: "setosa", 1: "versicolor", 2: "virginica"}
iris["target"] = np.vectorize(class_mapping.get)(iris["target"])  # class_mapping.get: 키 값 반환
iris
```

```python
df["Species"] = iris["target"] 
df
```

```python
df.drop("target", axis=1, inplace=True)
df.columns = ["SepalLength", "SepalWidth", "PetalLength", "PetalWidth", "Species"]
df
```

##### 1-3. 판다스 데이터 시각화
- 판다스의 시리즈나 데이터프레임은 'plot'이라는 시각화 메소드를 내장



```python
import matplotlib.pyplot as plt
df.plot(y="SepalLength", kind="bar", rot=0)
plt.title("SepalLength graph")
plt.xlabel("Data")
plt.ylabel("SepalLength")
plt.show()
```

```python
df.groupby("Species").mean()["SepalLength"].plot(kind="bar", rot=0)  # rot: 눈금, 제목 회전
plt.rc("font", family="Malgun Gothic")  # 글씨체 지정
plt.title("SepalLength mean")
plt.xlabel("Data")
plt.ylabel("Species")
plt.show()
```

```python
df.sort_values(by="Fare")
```

```python
df.drop("Cabin", axis=1, inplace=True)
df.head()
```

```python
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])  # 최빈값
df.isnull().sum()

# 동일한 일을 하는 코드

# # 최빈값 구하기
# df["Embarked"].value_counts()
# df["Embarked"] = df["Embarked"].fillna('S')
```

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
df.isnull().sum()
```

```python
# object 데이터로 그룹핑 X
# df.groupby("Embarked").mean()["Pclass"]

# 그룹핑이 가능한 열을 기준으로
df["Pclass"].groupby(df["Embarked"]).mean()
```

```python
df.drop(["Name", "Ticket", "Embarked"], axis=1).groupby(["Pclass", "Sex"]).mean()
```

##### 2-3. 시각화

```python
import seaborn as sns

sns.countplot(x="Pclass", hue="Survived", data=df)  # hue: 데이터 그룹화, 색상 부여, 범례 생성
```

```python
sns.heatmap(data=df.drop(["Name", "Ticket", "Embarked", "Sex"], axis=1).corr(), 
            annot=True, cmap="Blues") # annot: 각 셀에 상관관계 값 표시
```
