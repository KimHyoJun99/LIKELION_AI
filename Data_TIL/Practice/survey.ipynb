### Preprocessing

```python
# LanguageHaveWorkedWith

df_survey["LanguageHaveWorkedWith"] = df_survey["LanguageHaveWorkedWith"].str.split(';')
df_survey_explode = df_survey.explode("LanguageHaveWorkedWith")  # list 내의 요소들 풀어내기
df_survey_explode.head()
```

```python
one_hot = pd.get_dummies(df_survey_explode["LanguageHaveWorkedWith"])
df_survey_explode_1hot = pd.concat([df_survey_explode, one_hot], axis=1)
df_survey_explode_1hot.head()
```

```python
def count_lang(lang_li):
    global count_dict
    if (isinstance(lang_li, list)):  # NaN 값은 접근하지 않기 위해
        for lang in lang_li:
            if lang in count_dict:
                count_dict[lang] += 1
            else:
                count_dict[lang] = 1

count_dict = {}
df_survey["LanguageHaveWorkedWith"].apply(count_lang)

sorted_count_dict = dict(sorted(count_dict.items(), key=lambda item: item[1], reverse=True))  # 정렬

for lang, count in sorted_count_dict.items():
    print(lang, count)
```
